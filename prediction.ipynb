```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
### Load the ANN trained model, sclaer pickle file, onehot
model = load_model('model.h5')

##load the encoder and scaler
with open('label_encoder_gender.pkl', 'rb') as file:
    label_encoder_geo = pickle.load(file)

with  open('scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)
```

```python
# Example input data
input_data = {
'CreditScore': 600,
'Geography': 'France',
'Gender': 'Male',
'Age': 40,
'Tenure': 3,
'Balance': 60000,
'NumOfProducts': 2,
'HasCrCard': 1,
'IsActiveMember': 1,
'EstimatedSalary': 50000
}
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
onehot_encoder_geo = OneHotEncoder()

geo_encoded = onehot_encoder_geo.fit_transform(
    [[input_data['Geography']]]
).toarray()
```

```python
geo_encoded_df = pd.DataFrame(geo_encoded, columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_data = pd.DataFrame([input_data])
```

```python
## Combine one-hot encoded columns with input data
input_data = pd.concat(
    [input_data.reset_index(drop=True), geo_encoded_df],
    axis=1
)
```

```python
## Geography and Gender => Categorical Variables

# Gender -> 0 and 1
label_encoder_gender = LabelEncoder()
input_data['Gender'] = label_encoder_gender.fit_transform(input_data['Gender'])
```
