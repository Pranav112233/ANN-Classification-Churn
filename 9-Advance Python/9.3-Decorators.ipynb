**Decorators**

Decorators are a powerful and flexible feature in Python that allows you to modify the behavior of a function or class method. They are commonly used to add functionality to functions or methods without modifying their actual code. This lesson covers the basics of decorators, including how to create and use them.

** Function Copy**

-> we can copy one function to the another function

```python
def welcome():
    return "Welcome to the advanced Python Course"

wel = welcome
print(wel())
```

```python
wel = welcome() #Copy of the welcome function
print(wel)
del welcome
```

### ** Closures **

```python
## closures functions -> return type is always the sub-method
def main_welcome():
    msg = "Welcome"
    def sub_welcome_method():
        print("Welcome to the advanced python course")

    ## Accessing local variable inside the closure function
        print(msg)

        print("Please these concepts seriously!")
    return sub_welcome_method()
```

```python
main_welcome()
```

We can pass methods inside a method


```python
def main_welcome(func, lst):
    def sub_welcome_method():
        print("Welcome to the advanced python course")
        print(func(lst))   # FIXED
        print("Please take these concepts seriously!")
    return sub_welcome_method()

main_welcome(len, [1,2,3,4,5,6])
```

### Decorators

```python
### Decorator
def main_welcome(func):
    def sub_welcome_method():
        print("Welcome to the advanced python course")
        func()   
        print("Please take these concepts seriously!")
    return sub_welcome_method()
```

```python
def course_introduction():
    print("this is an advanced python course")

course_introduction()
```

```python
main_welcome(course_introduction)
```

```python
# Now, without calling the main_welcome function, can we display all the above ??
# Decorator used here

@main_welcome #decorator keyword
def course_introduction():
    print("this is an advanced python course")

#course_introduction()
```

```python
# Some more examples
def my_decorator (func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper
```

```python
@my_decorator

def say_hello():
    print('Hello')
```

```python
say_hello()
```

```python
## Decorators with arguments

def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print('Hello')
```

```python
say_hello()
```
