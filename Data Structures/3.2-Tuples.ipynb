```python
#Immutable Nature of Tuples

# tuples are immutable, elemts can't be changed once they are assigned

lst = [1,2,3,4,5]
print(lst)

lst[1] = "Krish"
print(lst)
```

```python
##     Tuple Methods

print(numbers.count(1))
print(numbers.index(3))
```

```python
## Packing and Unpacking Tuple

#Packing
packed_tuple = 1, "Hello", 3.14
print(packed_tuple)
```

```python
#Unpacking
a,b,c = packed_tuple

print(a)
print(b)
print(c)
```

```python
## Unpacking with *

numbers = (1,2,3,4,5,6)
first, *middle, last = numbers
print(first)
print(middle)
print(last)
```

```python
## Nested Tuple

lst = [[1,2,3,4], [6,7,8,9], [1,1,"Hello",3.14,"c"]]
lst[0]
```

```python
lst[0][2]
```

```python
#slicing
lst[0][0:3]
```

```python
# Nested Tuple -> Tuples inside tuple
nested_tuple = ((1,2,3), ("a","b","c"), (True,False))

## access the elements inside a tuple
print(nested_tuple[0])
```

```python
# iterating over nested tuples

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```
