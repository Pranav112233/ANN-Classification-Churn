```python
## Even / Odd
def even_or_odd(num):
    """ This function finds even or odd"""
    if num % 2 == 0:
        print("Even")
    else:
        print("Odd")
```

```python
## Call this function
even_or_odd(24)
```

```python
## Default Parameters

def greet(name="Guest"):
    print(f"Hello {name}")


greet("Pranav")
```

```python
## Variable Length Arguments
## Positional and Keywords Arguments


#Positional Argument
def print_numbers(*args):
    for number in args:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,"Krish")
# *krish is referring to all positionl arguments
```

```python
# Keywords Arguments
def print_details(**kwargs): #all parameters in form of (key,value) pairs
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="Krish",age="32",country="India")
```

```python
## Combining both of them
def print_details(*args, **kwargs):
    for  val in args:
        print(f"Positional Argument : {val}")

    for key, value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,5,"Krish",name="Krish",age="32",country="India")
```

```python
## Return Statements
def multiply(a,b):
    return a * b

multiply(2,3)
```

```python
## Return multiple parameters from a function
def multiply(a,b):
    return a * b,a

multiply(2,3)
```
