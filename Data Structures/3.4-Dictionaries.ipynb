```python
# Dictionary store values in key-value pairs.

#Creating Dictionaries

# in Dictionaries we use flower braces
empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict = dict()
empty_dict
```

```python
student = {"name":"Krish","age":32,"grade":24}
print(student)
print(type(student))
```

```python
## Error -> Key sd be unique(don't make keys with  the same name)
student = {"name":"Krish","age":32,"name":24}
print(student)
```

```python
## Accessing Dictionary Element
student = {"name":"Krish","age":32,"grade":'A'}
print(student)
```

```python
print(student['grade'])
print(student['age'])
```

```python
## Accessing value by using get() method provided by Dictionary
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name',"Not Available"))
```

```python
#Modifying Dictionary Elements
# Dictionaries are mutable -> you can  add, update or delete elements

print(student)
```

```python
student['age']=33 #updated value for a key
print(student)
student['address']='India' #added a new key and value
print(student)
```

```python
#Deleting a Key
del student['grade'] #delete a key-value pair

print(student)
```

```python
## Dictionary Methods

keys = student.keys() #get all the keys
print(keys)
values = student.values() #get all the values
print(values)

items = student.items() # get all key value pairs

print(items)
```

```python
## Shallow Copy
student_copy = student
print(student_copy)
print(student)
```

```python
student["name"] = "Krish1"
print(student)
print(student_copy)
```

```python

student_copy1 = student.copy() ## shallow copy
print(student_copy1)
print(student)
```

```python
student_copy1["name"] = "Krish2"
print(student)
print(student_copy1)
```

```python
student["name"] = "Krish3"
print(student_copy1)
print(student)
```

```python
# shallow copy provides a different memory location, so when we change one of the  variable it does not impact the other variable
```

```python
## Iterating over Dictionaries 
## You can use loops to iterate over dictionaries, keys,values or items

#iterating over keys

for keys in student.keys():
    print(keys)
```

```python
#iterate over values
for values in student.values():
    print(values)
```

```python
##Iterate over key-value pairs
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## Nested Dictionaries 
students = {
    "student1":{"name":"Krish", "age":32},
    "student2":{"name":"Peter", "age":30}
}
print(students)
```

```python
## Access nested dictionaries elements
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
## Iterating over nested dictionaries
for student_id, student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary Comprehension

#square b/w 1 to 10

squares = {x:x**2 for x in range(5)}
print(squares)
```

```python
## Conditional Dictionary Comprehension

evens = {x:x**2 for x in range(10) if x%2==0} #square of only even numbers
print(evens)
```

```python
#Practical Examples

# Use a dictionary to count the freq of elements in list

numbers = [1,2,2,3,3,3,4,4,4,4]
frequency = {count:numbers.count(4) for count in numbers}
print(frequency)
```

```python
numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}
for number in numbers:
    if number in frequency:
        frequency [number]+=1
        
    else:
        frequency [number]=1
print(frequency)
```

```python
## Merge 2 Dictionaries

dict1 = {"a":1, "b":2}
dict2 = {"b":3, "c":4}
merge_dict = {**dict1, **dict2}

print(merge_dict)
```
