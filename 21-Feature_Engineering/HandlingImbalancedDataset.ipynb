### Handling Imbalanced Dataset

1. Up Sampling
2. Down Sampling

```python
import numpy as np 
import pandas as pd 

#Set the random seed for reproductivity
np.random.seed(123)

#Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0, n_class_1
```

```python
## Create Dataframe with imbalanced dataset
class_0 = pd.DataFrame({
'feature_1': np.random.normal(loc=0, scale =1, size= n_class_0),
'feature 2': np.random.normal(loc=0, scale=1, size =n_class_0),
'target': [0] * n_class_0
})

class_1= pd.DataFrame({
'feature 1': np.random.normal(loc=2, scale=1, size= n_class_1),
'feature 2': np.random.normal(loc=2, scale=1, size =n_class_1),
'target': [1]* n_class_1
})
```

```python
df = pd.concat([class_0, class_1]).reset_index(drop=True)
```

```python
df.head()
```

```python
df['target'].value_counts()
```

### Up Sampling

```python
# Here we increase the number of 1's by some artificial points
df_minority = df[df['target'] == 1]
df_majority = df[df['target'] == 0]
```

```python
from sklearn.utils import resample
# create extra bullet points
df_minority_upsampled = resample(df_minority, replace=True,#Sample with replacement
        n_samples = len(df_majority),
        random_state=42
        )
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
```

```python
df_upsampled['target'].value_counts()
```

### Downsampling

```python
## Create Dataframe with imbalanced dataset
class_0 = pd.DataFrame({
'feature_1': np.random.normal(loc=0, scale =1, size= n_class_0),
'feature 2': np.random.normal(loc=0, scale=1, size =n_class_0),
'target': [0] * n_class_0
})

class_1= pd.DataFrame({
'feature 1': np.random.normal(loc=2, scale=1, size= n_class_1),
'feature 2': np.random.normal(loc=2, scale=1, size =n_class_1),
'target': [1]* n_class_1
})

df = pd.concat([class_0, class_1]).reset_index(drop=True)

#Check the class distribution
print(df['target'].value_counts())
```

```python
# Downsampling -> Reducing Datapoints
df_minority = df[df['target'] == 1]
df_majority = df[df['target'] == 0]
```

```python
from sklearn.utils import resample
# create extra bullet points
df_majority_downsampled = resample(df_majority, replace=False,#Sample with replacement
        n_samples = len(df_minority), #reduce datapoints to the length of minority
        random_state=42
        )
```

```python
df_majority_downsampled.shape
```

```python
df_downsampled = pd.concat([df_minority, df_majority_downsampled])
```

```python
df_downsampled.target.value_counts()
```
