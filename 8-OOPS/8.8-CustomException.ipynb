## Custome Exception (Raise and Throw an Exception)

```python
class Error(Exception):
    pass

class dobException(Error):
    pass
```

```python
year = int(input("Enter The Date of Birth"))
age = 2024 - year

try:
    if age <=30 and age >=20:
        print("Age is Valid So you can apply for the exams")
    else:
        raise  dobException
except dobException:
    print("Sorry, your age should be greater than 20 or less than 30")
```
