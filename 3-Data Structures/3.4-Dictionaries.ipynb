```python
# Dictionary store values in key-value pairs.

#Creating Dictionaries

# in Dictionaries we use flower braces
empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict = dict()
empty_dict
```

```python
student = {"name":"Krish","age":32,"grade":24}
print(student)
print(type(student))
```

```python
## Error -> Key sd be unique(don't make keys with  the same name)
student = {"name":"Krish","age":32,"name":24}
print(student)
```

```python
## Accessing Dictionary Element
student = {"name":"Krish","age":32,"grade":'A'}
print(student)
```

```python
print(student['grade'])
print(student['age'])
```

```python
## Accessing value by using get() method provided by Dictionary
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name',"Not Available"))
```

```python
#Modifying Dictionary Elements
# Dictionaries are mutable -> you can  add, update or delete elements

print(student)
```

```python
student['age']=33 #updated value for a key
print(student)
student['address']='India' #added a new key and value
print(student)
```

```python
#Deleting a Key
del student['grade'] #delete a key-value pair

print(student)
```

```python
## Dictionary Methods

keys = student.keys() #get all the keys
print(keys)
values = student.values() #get all the values
print(values)

items = student.items() # get all key value pairs

print(items)
```

```python
## Shallow Copy
student_copy = student
print(student_copy)
print(student)
```

```python
student["name"] = "Krish1"
print(student)
print(student_copy)
```

```python
# It depicts when doing some change in the original dictionary the copy of it will also be affected (Shallow copy).

student_copy1 = student.copy() ## shallow copy
print(student_copy1)
print(student)
```

```python
student["name"] = "Krish2"
print(student)
print(student_copy)
```
